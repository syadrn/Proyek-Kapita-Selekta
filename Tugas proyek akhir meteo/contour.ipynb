### JANUARI - MARET (KALIMANTAN)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load data
file_path = 'hasil.xlsx'
df = pd.read_excel(file_path)
df1= pd.read_excel(file_path,'ff_x')
df2 = pd.read_excel(file_path, 'ddd_x')

# Ekstrak data untuk Januari-Maret
date = '31-03-2023'
latitudes = df['latitude']
longitudes = df['longitude']
temperatures = df[date]
wind_speeds = df1[date]
wind_directions = df2[date]
wind_directions_rad = np.radians(wind_directions)
station_name = df['station_name']

# Konversi ke array numpy
lat = np.array(latitudes)
lon = np.array(longitudes)
tavg = np.array(temperatures)
wsx = np.array(wind_speeds)
wdx = np.array(wind_directions_rad)
station_names = np.array(station_name)

# Hapus data dengan NaN
valid_indices = ~np.isnan(tavg) & ~np.isnan(lat) & ~np.isnan(lon)
lat = lat[valid_indices]
lon = lon[valid_indices]
tavg = tavg[valid_indices]

u = wsx * np.sin(wdx)
v = wsx * np.cos(wdx)

# Membuat grid untuk interpolasi
grid_x, grid_y = np.mgrid[min(lon):max(lon):100j, min(lat):max(lat):100j]
grid_tavg = griddata((lon, lat), tavg, (grid_x, grid_y), method='linear')

# Plot data
plt.figure(figsize=(10, 8))
plt.quiver(lon, lat, u, v, angles='xy', scale_units='xy', scale=20, color='blue')
plt.pcolormesh(grid_x, grid_y, grid_tavg, shading='auto', cmap='coolwarm')
plt.colorbar(label='Temperature (Â°C)')
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(lon, lat, color = 'black', marker='o')
plt.title(f'Temperature Distribution for January-March 2023')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load data
file_path = 'hasil.xlsx'
df = pd.read_excel(file_path)
df1 = pd.read_excel(file_path, 'ff_x')  # Kecepatan angin
df2 = pd.read_excel(file_path, 'ddd_x')  # Arah angin

# Ekstrak data untuk Januari-Maret
date = '31-03-2023'
latitudes = df['latitude']
longitudes = df['longitude']
temperatures = df[date]
wind_speeds = df1[date]
wind_directions = df2[date]
wind_directions_rad = np.radians(wind_directions)  # Konversi arah angin ke radian
station_name = df['station_name']

# Konversi ke array numpy
lat = np.array(latitudes)
lon = np.array(longitudes)
tavg = np.array(temperatures)
wsx = np.array(wind_speeds)
wdx = np.array(wind_directions_rad)
station_names = np.array(station_name)

# Hapus data dengan NaN untuk suhu, kecepatan angin, dan arah angin
valid_indices = ~np.isnan(tavg) & ~np.isnan(lat) & ~np.isnan(lon) & ~np.isnan(wsx) & ~np.isnan(wdx)
lat = lat[valid_indices]
lon = lon[valid_indices]
tavg = tavg[valid_indices]
wsx = wsx[valid_indices]
wdx = wdx[valid_indices]

# Komponen vektor angin (u: arah timur-barat, v: arah utara-selatan)
u = wsx * np.sin(wdx)  # Komponen u (arah x)
v = wsx * np.cos(wdx)  # Komponen v (arah y)

# Membuat grid untuk interpolasi suhu
grid_x, grid_y = np.mgrid[min(lon):max(lon):100j, min(lat):max(lat):100j]
grid_tavg = griddata((lon, lat), tavg, (grid_x, grid_y), method='linear')

# Plot data suhu dan arah angin
plt.figure(figsize=(10, 8))

# Peta distribusi suhu
plt.pcolormesh(grid_x, grid_y, grid_tavg, shading='auto', cmap='coolwarm')
plt.colorbar(label='Temperature (Â°C)')
plt.gca().set_aspect('equal', adjustable='box')

# Plot vektor arah angin
plt.quiver(lon, lat, u, v, angles='xy', scale_units='xy', scale=50, color='blue', width=0.0025)

# Plot titik stasiun
plt.scatter(lon, lat, color='black', marker='o')

# Judul dan label sumbu
plt.title(f'Temperature and Wind Vectors for January-March 2023')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Tampilkan plot
plt.show()
```
