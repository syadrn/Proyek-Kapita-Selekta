```python
import pandas as pd
import glob

def process_file(file_path):
    # Read the specific column (C) and rows (3 to 5), with no header
    head = pd.read_excel(file_path, usecols="C", nrows=5, header=None)
    # Remove the head, replace double point and two spaces from each cell
    head.iloc[:, 0] = head.iloc[:, 0].str.replace(": ", "", regex=False)
    
    station_id = head.iloc[0, 0]  # Station ID
    station_name = head.iloc[1, 0]  # Station name
    latitude = float(head.iloc[2, 0])  # Latitude
    longitude = float(head.iloc[3, 0])  # Longitude
    elevation = float(head.iloc[4, 0])  # Elevation
    
    # Read data starting from row 9 without limiting the number of rows
    data = pd.read_excel(file_path, usecols="A:K", skiprows=8)
    
    # Find the index of the row containing "Keterangan:"
    keterangan_index = data[data.iloc[:, 0].str.contains("Keterangan :", na=False)].index
    if not keterangan_index.empty:
        last_row_index = keterangan_index[0] - 4  # rows before the "Keterangan" row
        relevant_data = data.loc[:last_row_index]
    else:
        relevant_data = data  # If "Keterangan" is not found, use all data
    
    # Transpose the data
    data_transposed = relevant_data.T
    # Extract the headers from row 1 (after transposition)
    headers = data_transposed.iloc[0].values

    # Extract tavg data from row 4 (Index 3 after transposition)
    tavg_data = data_transposed.iloc[3].values
    tavg_df = pd.DataFrame([tavg_data], columns=headers)

    # Extract ffavg data from row 5 (Index 4 after transposition)
    ffavg_data = data_transposed.iloc[4].values
    ffavg_df = pd.DataFrame([ffavg_data], columns=headers)

    # Extract ddd data from row 6 (Index 5 after transposition)
    ddd_data = data_transposed.iloc[5].values
    ddd_df = pd.DataFrame([ddd_data], columns=headers)

    metadata = {
        'station_id': [station_id],
        'station_name': [station_name],
        'latitude': [latitude],
        'longitude': [longitude],
        'elevation': [elevation]
    }
    metadata_df = pd.DataFrame(metadata)
    #Combine metadata with Tavg data
    tavg_df = pd.concat([metadata_df, tavg_df], axis=1)
    return tavg_df, ffavg_df, ddd_df

# Get a list of all relevant files
file_pattern = 'data_kalimantan/laporan_iklim_harian*.xlsx'
files = glob.glob(file_pattern)

# Initialize empty lists to hold DataFrames for each variable
all_tavg = []
all_ffavg = []
all_ddd = []

# Process each file and append the DataFrame to the corresponding list
for file in files:
    tavg_df, ffavg_df, ddd_df = process_file(file)
    all_tavg.append(tavg_df)
    all_ffavg.append(ffavg_df)
    all_ddd.append(ddd_df)

# Combine all the data for each variable
final_tavg_data = pd.concat(all_tavg, ignore_index=True)
final_ffavg_data = pd.concat(all_ffavg, ignore_index=True)
final_ddd_data = pd.concat(all_ddd, ignore_index=True)

# Save to Excel with 3 different sheets
output_file = 'data_kalimantan/hasil.xlsx'
with pd.ExcelWriter(output_file) as writer:
    final_tavg_data.to_excel(writer, sheet_name='TAVG', index=False)
    final_ffavg_data.to_excel(writer, sheet_name='FFAVG', index=False)
    final_ddd_data.to_excel(writer, sheet_name='DDD', index=False)

print(f"Output saved to {output_file}")
print(final_tavg_data)
```
