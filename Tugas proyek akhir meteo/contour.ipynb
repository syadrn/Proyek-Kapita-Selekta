### JANUARI - MARET (KALIMANTAN)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load data
file_path = 'hasil.xlsx'
df = pd.read_excel(file_path)


# Ekstrak data untuk Januari-Maret
date = '31-03-2023'
latitudes = df['latitude']
longitudes = df['longitude']
temperatures = df[date]
station_name = df['station_name']

# Konversi ke array numpy
lat = np.array(latitudes)
lon = np.array(longitudes)
tavg = np.array(temperatures)
station_names = np.array(station_name)

# Hapus data dengan NaN
valid_indices = ~np.isnan(tavg) & ~np.isnan(lat) & ~np.isnan(lon)
lat = lat[valid_indices]
lon = lon[valid_indices]
tavg = tavg[valid_indices]

# Membuat grid untuk interpolasi
grid_x, grid_y = np.mgrid[min(lon):max(lon):100j, min(lat):max(lat):100j]
grid_tavg = griddata((lon, lat), tavg, (grid_x, grid_y), method='linear')

# Plot data
plt.figure(figsize=(10, 8))
plt.pcolormesh(grid_x, grid_y, grid_tavg, shading='auto', cmap='coolwarm')
plt.colorbar(label='Temperature (°C)')
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(lon, lat, color = 'black', marker='o')
plt.title(f'Temperature Distribution for January-March 2023')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load data
file_path = 'data_kalimantan/hasil.xlsx'
# file_path2 = 'hasil2.xlsx'
df = pd.read_excel(file_path, 'TAVG')
# df1 = pd.read_excel(file_path2, 'FFAVG')
# df2 = pd.read_excel(file_path2, 'DDD')
df1 = pd.read_excel(file_path, 'FFAVG')
df2 = pd.read_excel(file_path, 'DDD')

# Ekstrak data untuk Januari-Maret
date = '31-03-2023'
latitudes = df['latitude']
longitudes = df['longitude']
temperatures = df[date]
wind_speeds = df1[date]
wind_directions = df2[date]
wind_directions_rad = np.radians(wind_directions)
station_name = df['station_name']

# Konversi ke array numpy
lat = np.array(latitudes)
lon = np.array(longitudes)
tavg = np.array(temperatures)
wsx = np.array(wind_speeds)
wdx = np.array(wind_directions_rad)
station_names = np.array(station_name)

# Hapus data dengan NaN di lat, lon, tavg, wsx, dan wdx
valid_indices = ~np.isnan(tavg) & ~np.isnan(lat) & ~np.isnan(lon) & ~np.isnan(wsx) & ~np.isnan(wdx)
lat = lat[valid_indices]
lon = lon[valid_indices]
tavg = tavg[valid_indices]
wsx = wsx[valid_indices]
wdx = wdx[valid_indices]

# Komponen kecepatan angin u dan v
u = wsx * np.sin(wdx)
v = wsx * np.cos(wdx)

# Membuat grid untuk interpolasi
grid_x, grid_y = np.mgrid[min(lon):max(lon):100j, min(lat):max(lat):100000j]
grid_tavg = griddata((lon, lat), tavg, (grid_x, grid_y), method='linear')

# Plot data
plt.figure(figsize=(10, 8))

# Plot distribusi temperatur (pcolormesh)
plt.pcolormesh(grid_x, grid_y, grid_tavg, shading='auto', cmap='coolwarm')
plt.colorbar(label='Temperature (°C)')
plt.gca().set_aspect('equal', adjustable='box')

# Plot vektor angin (quiver) di atas contour map
plt.quiver(lon, lat, u, v, angles='xy', scale_units='xy', scale=30, color='blue')

# Plot titik stasiun
plt.scatter(lon, lat, color='black', marker='o')

# Tambahkan judul dan label sumbu
plt.title(f'Temperature and Wind Distribution for January-March 2023')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Tampilkan plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import griddata
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load data
file_path = 'data_kalimantan/hasil.xlsx'
df = pd.read_excel(file_path, 'TAVG')
df1 = pd.read_excel(file_path, 'FFAVG')
df2 = pd.read_excel(file_path, 'DDD')

# Ekstrak data untuk Januari-Maret
date = '31-03-2023'
latitudes = df['latitude']
longitudes = df['longitude']
temperatures = df[date]
wind_speeds = df1[date]
wind_directions = df2[date]
wind_directions_rad = np.radians(wind_directions)
station_name = df['station_name']

# Konversi ke array numpy
lat = np.array(latitudes)
lon = np.array(longitudes)
tavg = np.array(temperatures)
wsx = np.array(wind_speeds)
wdx = np.array(wind_directions_rad)
station_names = np.array(station_name)

# Hapus data dengan NaN di lat, lon, tavg, wsx, dan wdx
valid_indices = ~np.isnan(tavg) & ~np.isnan(lat) & ~np.isnan(lon) & ~np.isnan(wsx) & ~np.isnan(wdx)
lat = lat[valid_indices]
lon = lon[valid_indices]
tavg = tavg[valid_indices]
wsx = wsx[valid_indices]
wdx = wdx[valid_indices]

# Komponen kecepatan angin u dan v
u = wsx * np.sin(wdx)
v = wsx * np.cos(wdx)

# Membuat grid untuk interpolasi
grid_x, grid_y = np.mgrid[min(lon):max(lon):100j, min(lat):max(lat):100j]
grid_tavg = griddata((lon, lat), tavg, (grid_x, grid_y), method='linear')

# Buat peta menggunakan Cartopy
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

# Menambahkan peta dunia
ax.coastlines(resolution='10m', linewidth=1)
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Batas untuk peta wilayah Kalimantan
ax.set_extent([108, 120, -5, 5], crs=ccrs.PlateCarree())  # [min lon, max lon, min lat, max lat]

# Plot distribusi temperatur (pcolormesh)
contour = ax.pcolormesh(grid_x, grid_y, grid_tavg, shading='auto', cmap='coolwarm', transform=ccrs.PlateCarree())
plt.colorbar(contour, ax=ax, label='Temperature (°C)')

# Plot vektor angin (quiver) di atas contour map
ax.quiver(lon, lat, u, v, transform=ccrs.PlateCarree(), angles='xy', scale_units='xy', scale=30, color='blue')

# Plot titik stasiun
ax.scatter(lon, lat, color='black', marker='o', transform=ccrs.PlateCarree())

# Tambahkan judul dan label
plt.title(f'Temperature and Wind Distribution for January-March 2023')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Tampilkan plot
plt.show()
```
